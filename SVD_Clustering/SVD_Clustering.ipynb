### Question 1: Single Value Decomposition (20 Points)
Use numpy to perform SVD on 'iris_features'. Run this experiment with rank 2 and 3 approximations. Write a function to compute the difference of Frobenius norm between the original matrix and your reconstructed matrix for each of these iterations and compare your results.

```python
import numpy as np
u, s, v_trans = np.linalg.svd(iris_features, full_matrices=False)
```

```python
def reduce_u(n,u):
    u_drop = u[:, :-n].copy()
    return u_drop
```

```python
def reduce_v(n,v):
    v_drop = v[:-n, :].copy()
    return v_drop
```

```python
def reduce_s(n,s):
    s_drop = s[:-n].copy()
    dim = len(s_drop)
    s_matrix = np.zeros((dim, dim))
    
    for i in range(dim):
        s_matrix[i][i] = s[i]
    
    return s_matrix
```

```python
n = 2
u_drop2 = reduce_u(n,u)
s_drop2 = reduce_s(n,s)
v_drop2 = reduce_v(n,v_trans)

svd2 = u_drop2 @ s_drop2 @ v_drop2

n = 1
u_drop3 = reduce_u(n,u)
s_drop3 = reduce_s(n,s)
v_drop3 = reduce_v(n,v_trans)
svd3 = u_drop3 @ s_drop3 @ v_drop3
```

```python
print("Original Matrix")
print(iris_features)
```

```python
print("Rank 2 Reduced Matrix")
print(svd2)
```

```python
print("Rank 3 Reduced Matrix")
print(svd3)
```

```python
def fro_norm(o,x):
    original = np.linalg.norm(o)
    other = np.linalg.norm(x)
    print("Original Matrix Frobenius Norm")
    print(original)
    print("Other Matrix Frobenius Norm")
    print(other)
    diff = original - other
    print("The Difference:" )
    print(diff)
```

```python
print("Rank 2 Matrix")
fro_norm(iris_features, svd2)
```

```python
print("Rank 3 matrix")
fro_norm(iris_features, svd3)
```

```python
#Since we used dimsionality reduction in the problem above, I will use the 3 dimensional version of Iris features. 
#The reason why I picked the rank 3 version is because the last dimension appears to have low significance.

def pts_to_xyz(some_data):
    x = []
    y = []
    z = []
    for p in some_data:
        x.append(p[0])
        y.append(p[1])
        z.append(p[2])
    return x, y, z


```

```python
wcss = []

for j in range(1,11):
    kmeans = KMeans(n_clusters=j).fit(df)
    print('number of clusters (k): {}'.format(j))
    print('cluster centers: ')
    print(kmeans.cluster_centers_)
    print()
    wcss.append(kmeans.inertia_)
```
